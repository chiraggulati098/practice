```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
driver = webdriver.Chrome() # Initialize Chrome WebDriver

driver.get("https://www.ycombinator.com/companies?batch=W24") # Go to the target web page

def scroll_down():
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(1)  # Wait for 1 second after scrolling (having no wait time was leading to some glitches)

stop_condition = False

while not stop_condition:
    scroll_down()
    page_source = driver.page_source # get HTML of the page after scrolling
    soup = BeautifulSoup(page_source, "lxml") # parse the webpage
    company_count_element = soup.find('div', {'class': '_message_99gj3_543'})
    # Check if stop condition is met (showing all companies)
    if company_count_element and "Showing 247 of 247 companies" in company_count_element.get_text():
        stop_condition = True
        print("All Companies Loaded")

driver.quit() # Close the WebDriver Window

soup = BeautifulSoup(page_source, "lxml") # Parse HTML (now all companies have been loaded)
```

```python
base_url = 'https://www.ycombinator.com'
```

```python
company_list = [] # Initialize an empty list
for a in (soup.find_all('a', {'class': '_company_99gj3_339'})):
    # Append tuple of (company name, company webpage)
    company_list.append((a.find('span', {'class': '_coName_99gj3_454'}).get_text(strip=True),base_url+a.get('href'))) 
```

```python
company_list #list of tuples of company name, url
```

```python
# Initialize empty lists
number_of_founders = []
team_size = []
company_location = []
```

```python
'''
Read all the company webpages and scrape them for getting the following details:
1. Number of Founders
2. Team Size
3. Company Location
and we will store all of these in the lists we just initiated
'''

i = 1
for company,company_url in company_list:
    print(i)
    i+=1
    response = requests.get(company_url)
    c_soup = BeautifulSoup(response.content, 'lxml')

    idx = 0
    if len(c_soup.find_all('div', {'class': 'flex flex-row justify-between'})) == 5:
        idx = 1
    number_of_founders.append(len(c_soup.find('div', {'class': 'space-y-5'})))
    team_size.append(int(c_soup.find_all('div', {'class': 'flex flex-row justify-between'})[1+idx].get_text()[10:].strip()))
    company_location.append(c_soup.find_all('div', {'class': 'flex flex-row justify-between'})[2+idx].get_text()[9:].strip())
```

```python
# Let's check the length of all the data structures (This should be same and be equal to 247 - which is the total number of companies)
print(len(company_list))
print(len(number_of_founders))
print(len(team_size))
print(len(company_location))
```

```python
# Created a Pandas DataFrame of all the details
scraped_data = pd.DataFrame({'Company Name': [item[0] for item in company_list],
                             'Number of Founders': number_of_founders,
                             'Team Size': team_size,
                             'Company Location': company_location
                            })
```

```python
scraped_data.head()
# With all this data, we can now create any graphs that we might want
```

```python
scraped_data['Company Location'].value_counts()
```

```python
scraped_data['Company Location'] = scraped_data['Company Location'].replace('', 'Unknown') 
# for some companies, the location wasn't mentioned, we will replace it to 'Unknown'
```

```python
scraped_data['Company Location'].value_counts()
```

```python
sns.set_style("whitegrid") # set style
```

```python
# Graph for Location x Companies

plt.figure(figsize=(12, 6))
sns.countplot(x='Company Location', data=scraped_data, palette='viridis')

plt.xlabel('Company Location')
plt.ylabel('Number of Companies')
plt.title('Location of Companies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Graph for Team Size x Companies

plt.figure(figsize=(12, 6))
sns.countplot(x='Team Size', data=scraped_data, palette='viridis')

plt.xlabel('Team Size')
plt.ylabel('Number of Companies')
plt.title('Team Size')
plt.tight_layout()
plt.show()
```
