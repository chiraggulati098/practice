```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
l1 = [x for x in (1,11)]
l2 = [x**2 for x in l1]
```

```python
plt.plot(l1,l2)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_title('Title')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('BIG')
axes2.set_title('SMALL')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,1)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.3,0.5,0.3,0.3])
axes1.plot(x,y)
axes2.plot(y,x,'r')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

ax[0].plot(x,y)
ax[1].plot(x,y)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()
axis = fig.add_axes([0,0,1,1])

axis.plot(x,x**2,label='X2')
axis.plot(x,x**3,label='X3')

axis.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple',lw = 3,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
