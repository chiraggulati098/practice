#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
vertices = np.array([[400,400],[200,700],[600,700]])
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
new_img = img.copy()
cv2.polylines(new_img,[pts],isClosed=True,color=(0,0,255),thickness=10)
plt.imshow(new_img)
```

#### TASK: (NOTE: YOU WILL NEED TO RUN THIS AS A SCRIPT). Create a script that opens the picture and allows you to draw empty red circles whever you click the RIGHT MOUSE BUTTON DOWN.

```python
import numpy as np
import cv2

img = cv2.imread('../DATA/dog_backpack.jpg')
img = cv2.resize(img,(0,0),img,0.65,0.65)

def draw_circle(event,x,y,flag,params):

    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),radius=35,color=(0,0,255),thickness=5)

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)

while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
